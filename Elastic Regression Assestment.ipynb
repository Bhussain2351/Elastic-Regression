Q3. What are the advantages and disadvantages of Elastic Net Regression?

Ans. 
Advantages of Elastic Net Regression
Handles Multicollinearity:

Unlike Lasso, Elastic Net can handle highly correlated predictors effectively by selecting groups of correlated features instead of arbitrarily dropping some.
Combines L1 and L2 Regularization:

It benefits from both:
L1 (Lasso): For feature selection and sparsity.
L2 (Ridge): For stability and handling multicollinearity.
Feature Selection:

Similar to Lasso, Elastic Net can shrink some coefficients to zero, effectively selecting the most important features.
